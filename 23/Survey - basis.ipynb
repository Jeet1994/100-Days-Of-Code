```python
import pandas as pd
import numpy as np
```

```python
# Define the distributions based on the given percentages (normalized if necessary)
n_participants = 295
```

```python
# Gender
gender_dist = [0.5627, 0.4373]
genders = np.random.choice(['Male', 'Female'], n_participants, p=gender_dist)
```

```python
# Age
age_dist = np.array([0.3220, 0.4237, 0.1525, 0.1017])
age_dist /= age_dist.sum()
ages = np.random.choice(['Younger than 30', '30 – 39', '40 – 49', 'Above 50'], n_participants, p=age_dist)
```

```python
# Occupation
occupation_dist = [0.3220, 0.4102, 0.1763, 0.0847, 0.0068]
occupations = np.random.choice(['Teaching', 'Engineering', 'Medical', 'Govt.', 'Others'], n_participants, p=occupation_dist)
```

```python
# Education
education_dist = np.array([0.0320, 0.1030, 0.0710, 0.6350, 0.1600])
education_dist /= education_dist.sum()
education_levels = np.random.choice(['Below High School', 'High School', 'Diploma', 'Bachelor’s', 'Postgraduate'], n_participants, p=education_dist)

# Income
income_dist = [0.3424, 0.4373, 0.2203]
incomes = np.random.choice(['> 1000$ & < 5000 $', '> 5000$ & < 10000$', '> 10000$'], n_participants, p=income_dist)

# Expense
expense_dist = np.array([0.1627, 0.7186, 0.1186])
expense_dist /= expense_dist.sum()
expenses = np.random.choice(['> 1000$ & < 5000 $', '> 5000$ & < 10000$', '> 10000$'], n_participants, p=expense_dist)

# Existing Ailment
ailment_dist = [0.38, 0.49, 0.13]
ailments = np.random.choice(['Yes', 'No', 'Cannot disclose'], n_participants, p=ailment_dist)
```

```python
# Combine into a DataFrame
synthetic_data = pd.DataFrame({
    'Gender': genders,
    'Age': ages,
    'Occupation': occupations,
    'Education': education_levels,
    'Income': incomes,
    'Expense': expenses,
    'Existing Ailment': ailments
})

print(synthetic_data.head())
print(synthetic_data.describe(include='all'))
```
