```python
import pandas as pd
import numpy as np
```

```python
# Number of participants
n_participants = 295
```

```python

# Demographics distributions
gender_dist = [0.5627, 0.4373]
age_dist = np.array([0.3220, 0.4237, 0.1525, 0.1017])
age_dist /= age_dist.sum()
occupation_dist = [0.3220, 0.4102, 0.1763, 0.0847, 0.0068]
education_dist = np.array([0.0320, 0.1030, 0.0710, 0.6350, 0.1600])
education_dist /= education_dist.sum()
income_dist = [0.3424, 0.4373, 0.2203]
expense_dist = np.array([0.1627, 0.7186, 0.1186])
expense_dist /= expense_dist.sum()
ailment_dist = [0.38, 0.49, 0.13]
cities = ['Riyadh', 'Jeddah', 'Mecca', 'Medina', 'Dammam', 'Khobar', 'Tabuk', 'Abha', 'Hofuf', 'Taif']
city_dist = [0.25, 0.2, 0.15, 0.1, 0.1, 0.05, 0.05, 0.05, 0.025, 0.025]
```

```python
# Generate demographics data
genders = np.random.choice(['Male', 'Female'], n_participants, p=gender_dist)
ages = np.random.choice(['Younger than 30', '30 – 39', '40 – 49', 'Above 50'], n_participants, p=age_dist)
occupations = np.random.choice(['Teaching', 'Engineering', 'Medical', 'Govt.', 'Others'], n_participants, p=occupation_dist)
education_levels = np.random.choice(['Below High School', 'High School', 'Diploma', 'Bachelor’s', 'Postgraduate'], n_participants, p=education_dist)
incomes = np.random.choice(['> 1000$ & < 5000 $', '> 5000$ & < 10000$', '> 10000$'], n_participants, p=income_dist)
expenses = np.random.choice(['> 1000$ & < 5000 $', '> 5000$ & < 10000$', '> 10000$'], n_participants, p=expense_dist)
ailments = np.random.choice(['Yes', 'No', 'Cannot disclose'], n_participants, p=ailment_dist)
locations = np.random.choice(cities, n_participants, p=city_dist)
```

```python

# Generate survey responses
product_satisfaction = np.random.choice([1, 2, 3, 4, 5], n_participants)
usage_frequency = np.random.choice(['Daily', 'Weekly', 'Monthly', 'Rarely', 'Never'], n_participants)
customer_service_rating = np.random.choice([1, 2, 3, 4, 5], n_participants)
support_speed = np.random.choice(['Very quickly', 'Quickly', 'Moderately', 'Slowly', 'Very slowly'], n_participants)
improvements = np.random.choice(['Better quality', 'More features', 'Lower price', 'Better customer service', 'Other'], n_participants)

# Additional survey questions (Yes/No)
medical_care = np.random.choice(['Yes', 'No'], n_participants)
easier_health_management = np.random.choice(['Yes', 'No'], n_participants)
safer_daily_life = np.random.choice(['Yes', 'No'], n_participants)
better_life_with_tracker = np.random.choice(['Yes', 'No'], n_participants)
easy_to_use_monitor = np.random.choice(['Yes', 'No'], n_participants)
easy_learning = np.random.choice(['Yes', 'No'], n_participants)
convenient_to_use = np.random.choice(['Yes', 'No'], n_participants)
family_approval = np.random.choice(['Yes', 'No'], n_participants)
friends_recommend = np.random.choice(['Yes', 'No'], n_participants)
friends_approval = np.random.choice(['Yes', 'No'], n_participants)
know_how_to_use = np.random.choice(['Yes', 'No'], n_participants)
help_available = np.random.choice(['Yes', 'No'], n_participants)
sufficient_finances = np.random.choice(['Yes', 'No'], n_participants)
positive_impact = np.random.choice(['Yes', 'No'], n_participants)
benefit_family_life = np.random.choice(['Yes', 'No'], n_participants)
feel_positive = np.random.choice(['Yes', 'No'], n_participants)
happy_to_use = np.random.choice(['Yes', 'No'], n_participants)
better_mental_health = np.random.choice(['Yes', 'No'], n_participants)
increase_quality_of_life = np.random.choice(['Yes', 'No'], n_participants)

# Combine into a DataFrame
synthetic_data = pd.DataFrame({
    'Gender': genders,
    'Age': ages,
    'Occupation': occupations,
    'Education': education_levels,
    'Income': incomes,
    'Expense': expenses,
    'Existing Ailment': ailments,
    'Location': locations,
    'Product Satisfaction': product_satisfaction,
    'Usage Frequency': usage_frequency,
    'Customer Service Rating': customer_service_rating,
    'Support Speed': support_speed,
    'Improvement Suggestions': improvements,
    'Medical Care': medical_care,
    'Managing Health Easier': easier_health_management,
    'Safer Daily Life': safer_daily_life,
    'Better Life with Tracker': better_life_with_tracker,
    'Easy to Use Monitor': easy_to_use_monitor,
    'Easy Learning': easy_learning,
    'Convenient to Use': convenient_to_use,
    'Family Approval': family_approval,
    'Friends Recommend': friends_recommend,
    'Friends Approval': friends_approval,
    'Know How to Use': know_how_to_use,
    'Help Available': help_available,
    'Sufficient Finances': sufficient_finances,
    'Positive Impact': positive_impact,
    'Benefit Family Life': benefit_family_life,
    'Feel Positive': feel_positive,
    'Happy to Use': happy_to_use,
    'Better Mental Health': better_mental_health,
    'Increase Quality of Life': increase_quality_of_life
})

print(synthetic_data.head())
print(synthetic_data.describe(include='all'))
```
